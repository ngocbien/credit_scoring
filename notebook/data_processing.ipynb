```python
not_nulls = []
for col in df.columns:
    if df[col].isnull().sum()==0:
        not_nulls.append(col)
print(not_nulls)
```

```python
len(not_nulls)
f, axs = plt.subplots(3,4,figsize=(15,15))
i = 0
for col in axs:
    for row in col:
        name = not_nulls[i]
        t = df[name].value_counts(dropna=False)
        x1 = t.plot.bar(ax=row)
        #x2 = df[df.label==1][name].value_counts(dropna=False).plot.bar(ax=row)
        plt.sca(row)
        if len(t)<10:
            plt.xticks(range(len(t)), t.index, rotation=90)
        else:
            plt.xticks(range(0,len(t),3), t.index[range(0,len(t),3)], rotation=90)
        row.set_title(name)
        i+=1
        if i>=len(not_nulls):
            break
plt.show()
```
